```python
from prophet import Prophet
```

```python
grp=retail[['Date','Orders']]
grp=grp.rename(columns={'Date':'ds','Orders':'y'})
grp
```

```python
php=Prophet()
php.fit(grp)
```

```python
future = php.make_future_dataframe(periods=90)
forecast = php.predict(future)

forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].to_csv("swiggy_forecast_output.csv", index=False)
```
